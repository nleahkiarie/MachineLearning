```python
import pandas as pd
import numpy as np
```

```python
data_df=pd.read_csv('/home/leah/Downloads/realEstate.csv') 
```

```python
data_df.head()
```

```python
#calculate correlation for all pairwise combinations
data_df['X3 distance to the nearest MRT station'].corr(data_df['X4 number of convenience stores'])
```

```python
x=data_df.drop(['Y house price of unit area',],axis=1).values
y=data_df['Y house price of unit area'].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(x, y)
```

```python
print('intercept:', model.intercept_)
```

```python
print('slope:', model.coef_)
```

```python
y_pred = model.predict(x)
print('predicted response:', y_pred, sep='\n')
```

```python
x_new = np.arange(35).reshape((-1, 7))
print(x_new)
```

```python
y_new = model.predict(x_new)
print(y_new)
```

```python
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
```
