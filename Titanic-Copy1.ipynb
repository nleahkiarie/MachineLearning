```python
#used to convert the string vaue to an integer place holder
from sklearn.preprocessing import LabelEncoder
```

```python
#encode the sex and cabin datatypes to integer in order to model with
encoder = LabelEncoder()
df.Sex = encoder.fit_transform(df.Sex)
df.Cabin = encoder.fit_transform(df.Cabin)

  
```

```python
#split the data into input and output datasets
```

```python
Input=df.drop(['Survived','Fare','Pclass','Embarked','Ticket','SibSp','Parch','Name'], axis='columns')
Result=df['Survived'].values
```

```python
#Split the data into training and testing
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(Input, Result,
test_size=0.3, random_state=1)
print(y_test)
```

```python
#use decision tree classifier and graphviz to model the decision tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
tree_model = DecisionTreeClassifier(max_depth=4, random_state=1)
tree_model
```

```python
tree_model.fit(X_train, y_train)
```

```python
print("The coefficient of determination is %.10f" % tree_model.score(Input, Result))
```

```python
tree_model.predict(Input)
```

```python
export_graphviz(tree_model, out_file="assignment2_decision_tree.dot", feature_names=X_train.columns, impurity=False, filled=True)
```

```python
#to generate the decision tree
from sklearn import tree
```

```python
fig = plt.figure(figsize=(19, 8))
tree.plot_tree(tree_model, feature_names=X_train.columns, impurity=False, filled=True)
```

```python
fig.savefig("assignment2_decision_tree2.png")
```
